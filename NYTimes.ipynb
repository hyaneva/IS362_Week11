```python
import requests
import pandas as pd

#Insert you NYT API key
API_KEY = 'INSERT YOUR API KEY HERE'

#Defines the URL for the Top Stories API (using the 'home' section)
url = f'https://api.nytimes.com/svc/topstories/v2/home.json?api-key={API_KEY}'

#Send the request to the API
response = requests.get(url)

#Check if the request was successful
if response.status_code == 200:
    #Load the JSON data
    data = response.json()
    
    #Extract the list of articles (each article is a dictionary)
    articles = data.get('results', [])
    
    #Create a DataFrame from the list of articles
    df = pd.DataFrame(articles)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```
